```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('googleplaystore.csv')
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
print(df.describe())
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Installs'], bins=20, color='lightgreen')
plt.xlabel('Installs')
plt.ylabel('Frequency')
plt.title('Distribution of Installs')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Reviews'], df['Rating'], alpha=0.5, color='purple')
plt.xlabel('Reviews')
plt.ylabel('Rating')
plt.title('Scatter Plot: Reviews vs. Rating')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df['Category'].value_counts().plot(kind='bar', color='coral')
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Number of Apps per Category')
plt.xticks(rotation=90)
plt.show()
```
